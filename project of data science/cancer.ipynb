```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
row_data=pd.read_csv('cancer.csv')
data=row_data.copy()
```

```python
data.head()
```

```python
data=data.drop(columns='Unnamed: 32',axis=1)
```

```python
data.head()
```

```python
sns.countplot(data['diagnosis'],label='count')
```

```python
data=data.drop(columns='id',axis=1)
```

```python
data.head()
```

```python
x=data.drop(columns='diagnosis',axis=1)
y=data['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x_1=LabelEncoder()
y=labelencoder_x_1.fit_transform(y)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=445)
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train,x_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# addind input and first hidden layer
classifier=Sequential()
classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu',input_dim=30))
# addind second hidden layer
classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu'))
# adding the output layer
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer="Adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train,batch_size=100,epochs=150)
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.savefig('h.png')
```
